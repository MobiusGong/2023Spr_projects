# Relationship between US-China Flights and Foreign Direct Investment (FDI)

### What is Foreign Direct Investment (FDI)?
Foreign Direct Investment (FDI) refers to an investment made by a company or individual from one country into business interests or assets in another country.
### What can FDI indicates? 
FDI can provide insights into the relationships between the US and China. By examining FDI flows between the two countries, we can identify how historical events can affect the relationship between the two countries. 
### What can US-China flights indicates?
Examining the flows of US-China flights provides insights for not only the economic ties but also tourism and cultural exchange, corporate relationships, educational exchange, and political relations

## FDI between US and China

```python
from functions_definition import *
df_FDI_combined = read_and_combine_FDI_data('China to US.csv', 'US to China.csv')
plot_FDI_data(df_FDI_combined, 'CN to US in billion U.S. dollars', 'US to CN in billion U.S. dollars')
```

- The FDI flows from US to China is larger than from China to US
### The flow from US to China
- 2007~2008 shows a rapid increase which can be attributed to the growing Chinese economy. The 2008 Beijing Olympics could have contributed to this increase.
- 2008~2013 the increase stopped. This can be attributed to the global financial crisis that started in 2008 (https://hbr.org/2013/11/what-weve-learned-from-the-financial-crisis)
- 2013-2017 the increase starts again. It can be attributed to the recovery of the global economy and China's Belt and Road Initiative, announced in 2013
- 2017~2019 the stagnation can be attributed to the US-China trade war and the political tensions.
- 2019~2021 the gradual inscrease can be attributed to the Phase One Trade Deal between US and China In January 2020. Additionally, the COVID-19 pandemic exposed vulnerabilities in global supply chains, prompting some companies to invest in China to diversify their supply chains and improve resilience.
### The flow from China to US
- 2007~2015 the gradual liberalization of China's outward investment policies combined with the growing interest of Chinese companies in the US market.
- 2015~2016 China experienced a stock market crash in 2015, which prompted Chinese investors to seek more stable investments overseas (https://www.vox.com/2015/7/8/8908765/chinas-stock-market-crash-explained)
- 2016~2021 2016 South China sea dispute; 2018 US-China trade war (https://www.theguardian.com/news/2016/jul/12/south-china-sea-dispute-what-you-need-to-know-about-the-hague-court-ruling)
